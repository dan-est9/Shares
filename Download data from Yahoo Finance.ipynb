```python
#Installing Packages

import pandas as pd
import datetime as dt
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
from pandas_datareader import data as pdr
```

```python
#Time interval & Stocks

start_date = '2020-01-01'
end_date = '2025-01-31'
tickers  = ['AAPL','MSFT','NVDA','AMZN','GOOGL','2222.SR','META','TSLA','AVGO','TSM']
```

```python
#Download the information from yahoo finance. Focus on the Close Price of each share 

data = yf.download(tickers, start=start_date, end = end_date)['Close']
data.head()
```

```python
#replace de NaN values

data.ffill(inplace= True)

print("\nDatos después de reemplazar NaN:")
print(data.head(10))
```

```python
data.loc['2020-01-01'] = data.loc['2020-01-01'].fillna(0)
```

```python
data.head(10)
```

```python
pd.options.plotting.backend = 'matplotlib'
data.plot(figsize=(12,8))
plt.title("Close price shares", fontdict={
    'fontsize':25,
    'fontweight': 'bold',
    'color':'darkgreen',
    'fontfamily': 'Courier New'
})
```

```python
#Daily Returns

daily_returns = data.pct_change()
print("Retornos Diarios")
print(daily_returns.head())
```

```python
daily_returns.replace([np.inf, -np.inf], np.nan, inplace=True)
daily_returns.dropna(how='any', inplace=True)
```

```python
pd.options.plotting.backend = 'matplotlib'
daily_returns.plot(figsize=(12,8))
plt.title("Daily Returns", fontdict={
    'fontsize':25,
    'fontweight': 'bold',
    'color':'darkgreen',
    'fontfamily': 'Courier New'
})
```
